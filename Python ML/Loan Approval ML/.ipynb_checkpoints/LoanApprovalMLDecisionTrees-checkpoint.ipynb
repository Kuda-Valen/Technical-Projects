```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
# For visualizing the tree
from sklearn.tree import export_graphviz
import graphviz 

# --- 1. Load Data (Starting Point) ---
df = pd.read_csv("/Users/cashcrusaders/Desktop/Technical Projects/Python ML/Loan Approval ML/loan_approval_dataset.csv") 

# --- 2. Data Cleaning and Preparation ---

# a. Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# b. Drop the non-predictive 'loan_id' column
df = df.drop('loan_id', axis=1)

# c. Encode the Target Variable ('loan_status')
df['loan_status'] = df['loan_status'].replace({' Approved': 1, ' Rejected': 0})

# d. One-Hot Encode Categorical Features
# This creates new numerical columns for 'education' and 'self_employed'
categorical_cols = ['education', 'self_employed']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("Shape of the prepared data (rows, columns):", df_encoded.shape)
print("-" * 50)


# --- 3. Split Data into Features (X) and Target (y) ---

# X contains all columns except the target
X = df_encoded.drop('loan_status', axis=1) 
# y is the target column
y = df_encoded['loan_status'] 

# Split the data into Training (70%) and Testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3,    # 30% of data for testing
    random_state=42,  # Ensures the split is the same every time
    stratify=y        # Ensures the target proportions are maintained in both splits
)

print("Training set size:", X_train.shape[0])
print("Testing set size:", X_test.shape[0])
print("-" * 50)


# --- 4. Train the Decision Tree Model ---

# Initialize the Decision Tree Classifier
# We set 'max_depth' to prevent extreme overfitting (a good practice)
dtc = DecisionTreeClassifier(
    max_depth=5, 
    random_state=42
)

# Train the model using the training data
dtc.fit(X_train, y_train)
print("Decision Tree Model trained successfully.")
print("-" * 50)


# --- 5. Evaluate the Model ---

# Make predictions on the test set
y_pred = dtc.predict(X_test)

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy*100:.2f}%")

# Print Classification Report (gives precision, recall, f1-score)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Rejected (0)', 'Approved (1)']))
print("-" * 50)


# --- 6. Visualize the Decision Tree (Optional but highly recommended) ---
# NOTE: This step requires graphviz to be installed on your system

dot_data = export_graphviz(
    dtc, 
    out_file=None,
    feature_names=X.columns,
    class_names=['Rejected', 'Approved'],
    filled=True,
    rounded=True,
    special_characters=True
)

graph = graphviz.Source(dot_data)
graph.render("loan_approval_tree", format="png", view=True) # Uncomment to save and view the tree image

print("Decision Tree structure successfully exported to graphviz object.")

print("\n\nUsing the system----\n")

no_of_dependents = int(input("How dependencies do you have: "))
education = input("What is your education status [Graduate / Not Graduate]: ")
self_employed = input("Are you self employed [Yes / No]: ")
income_annum = 


user_data = {
    "no_of_dependents" = no_of_dependents,
    "education" = education,
    "self_employed" = self_employed,
    "income_annum" = income_annum,
    "loan_amount" = loan_amount,
    "loan_term" = loan_term,
    "cibil_score" = cibil_score,
    "residential_assets_value" = residential_assets_value,
    "commercial_assets_value" = commercial_assets_value,
    "luxury_assets_value" = luxury_assets_value,
    "bank_asset_value" = bank_asset_value,
}
```
