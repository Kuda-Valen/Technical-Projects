```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = ([[4],[5], [6], [8], [10]])
y = ([65, 70, 75, 85, 95])

model = LinearRegression()
model.fit(X, y)

m = model.coef_[0]
b = model.intercept_

print(f"The calculated slope(m) is: {m:.2f}")
print(f"The calculated y-intercept (b) is: {b:.2f}\n")

hours = int(input("\nEnter a hours to predict possible exam score: "))
x_2d = np.array([[hours]])
predicted_y = model.predict(x_2d)

print(f"The predicted exam score with {hours }hours studied is: {predicted_y}marks")

```
