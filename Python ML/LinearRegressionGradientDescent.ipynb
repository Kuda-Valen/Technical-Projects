```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Sample Data (X = years, y = price)
X = np.array([15, 20, 25, 30, 35])
y = np.array([250, 320, 380, 450, 500])

# Step 2: Initialize parameters
m = 0  # slope
b = 0  # intercept
alpha = 0.0001  # learning rate
epochs = 1000  # number of iterations
n = len(X)

# Step 3: Gradient Descent Loop
for i in range(epochs):
    y_pred = m * X + b  # prediction
    cost = (1/n) * sum((y - y_pred)**2)  # MSE

    # Compute gradients
    dm = (-2/n) * sum(X * (y - y_pred))
    db = (-2/n) * sum(y - y_pred)

    # Update parameters
    m = m - alpha * dm
    b = b - alpha * db

    # Optional: track progress
    if i % 100 == 0:
        print(f"Epoch {i}, Cost: {cost:.4f}, m: {m:.4f}, b: {b:.4f}")

# Step 4: Plot the final regression line
plt.scatter(X, y, color='red')
plt.plot(X, m*X + b, color='blue')
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("Years")
plt.ylabel("price")
plt.show()

name = input("What is your name: ")
option = 0
try: 
    while option != 4

    print("Predicting your House Prices according to Square feet......\n")
    print(f"Hello, {name}\n")
    print("1. View the ML details")
    print("2. Use the ML to make a price prediction")
    print("3. See the ML visually with the regression line")
    print("4. Exit")

    option = int(input("\nChoose an option: "))

    if option == 1:
        

except: 
    print("Invalid input, please enter valid input!")
```
