```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 4, 2, 5, 6])


model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
b = model.intercept_
m = model.coef_

x_input = int(input("Enter the X-input: "))
y_variable = model.predict([[x_input]])

print(f"\nWith your x-input {x_input}, the prediction is: {y_variable}")
print(f"The intercept is: {b}")
print(f"The slope m is: {m}")


# Visualization

plt.scatter(X, y, color='red', label='Data points')
plt.plot(X, y_pred, color='blue', label='Regression line')

# Highlight your input prediction
plt.scatter(x_input, y_variable, color='green', s=100, label='Your Prediction')

plt.title("Linear Regression using sklearn")
plt.xlabel("X-input")
plt.ylabel("y-outputs")
plt.legend()
plt.show()
```
