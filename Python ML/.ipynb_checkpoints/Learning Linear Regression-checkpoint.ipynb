```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[150], [160], [170], [180], [190]])
y = np.array([55, 65, 70, 78, 85])

model = LinearRegression()
model.fit(X, y)

slope_m = model.coef_[0]
y_intercept_b = model.intercept_

print(f"The calculated slope (m) is: {slope_m:.2f}")
print(f"The calculated y-intercept (b) is: {y_intercept_b:.2f}")

print()
print()

cm = int(input("Enter hieght of new person you want to calculate their height: "))
cm_2d = np.array([[cm]])
kg = model.predict(cm_2d)
print(f"The Predicted weight of someone who is {cm}cm tall is: {kg}kg")
```
